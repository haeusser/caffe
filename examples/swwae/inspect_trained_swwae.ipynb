```python
# imports and stuff
import os
os.chdir('../..')  # we're in $CAFFE_ROOT/examples/swwae
import sys
sys.path.insert(0, './python')
import caffe
from google.protobuf import text_format
import caffe.draw
from caffe.proto import caffe_pb2
from pylab import *
%matplotlib inline
```

```python
# set up solver
caffe.set_device(0)
caffe.set_mode_gpu()
net = caffe.Net('examples/swwae/train.prototxt',
                'examples/swwae/modelfiles/mnist_swwae_iter_4900.caffemodel',
                caffe.TEST)
```

```python
# shorthand for showing feature maps
def print_map(blob_name, c=0, cmap='gray'):
    figure(figsize=tuple(map(lambda x: max(x/12,0.5), net.blobs[blob_name].data[0,c].shape)))
    imshow(net.blobs[blob_name].data[0,c], cmap)
    title(blob_name + ' (c=' + str(c) +')')
    matplotlib.pyplot.axis('off')
```

```python
# forward pass
out = net.forward()
```
