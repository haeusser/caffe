- First, import the plotting tool. Adjust the paths so that they work for you. If you've added PyCNN to your sys.path already, this should work out of the box.

All you need to do is to specify the directory in which the log_db.db is located. Then instantiate a plotter.

```python
loc = '../test/mnist/log'  # this file is expected to be in pymill/examples
plotter = mp.MillPlot(location=loc)
```

Before you can plot anything, the plotter needs to query data. You can tell it to do so by calling
`fetch_range(it_min=0, it_max=None)`
which will fetch all data if no args are provided or
`fetch_last(it_count=1)`
which will fetch the last it_count entries.
For now, let's fetch all entries and plot the learning rate.

```python
plotter.fetch_range()
plotter.plot_lr()
```

In order to plot all outputs, use `plot()`

```python
plotter.plot()
```

Of course you can also look into the guts of the fetched data:

```python
plotter.cache_log_test
```
